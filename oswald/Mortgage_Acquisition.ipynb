# Mortgage Acquisition

```python
import pandas as pd
import numpy as np
```

# Load Data

```python
#Acquisition column names
acq_colms= ["LoanIdentifier","OriginationChannel", "SellerName", "OriginalInterestRate",
            "OriginalUPB", "OriginalLoanTerm", "OriginationDate", "FirstPaymentDate",
            "OriginalLoanToValueLTV","OriginalCombinedLoanToValueCLTV", "NumberofBorrowers", "OriginalDebtToIncomeRatio",
            "BorrowerCreditScoreAtOrigination", "FirstTimeHomeBuyerIndicator", "LoanPurpose",
            "PropertyType", "NumberOfUnits", "OccupancyType", "PropertyState", "ZipCodeShort",
            "PrimaryMortgageInsurancePercent", "ProductType", "CoBorrowerCreditScoreAtOrigination",
            "MortgageInsuranceType", "RelocationMortgageIndicator"];
```

```python
data2015Q1= pd.read_csv('RawData/FannieMae_SingleFamily/Acquisition_2015Q1.txt', sep="|", header=None, names=acq_colms)
data2015Q2= pd.read_csv('RawData/FannieMae_SingleFamily/Acquisition_2015Q2.txt', sep="|", header=None, names=acq_colms)
data2015Q3= pd.read_csv('RawData/FannieMae_SingleFamily/Acquisition_2015Q3.txt', sep="|", header=None, names=acq_colms)
data2015Q4= pd.read_csv('RawData/FannieMae_SingleFamily/Acquisition_2015Q4.txt', sep="|", header=None, names=acq_colms)
data2016Q1= pd.read_csv('RawData/FannieMae_SingleFamily/Acquisition_2016Q1.txt', sep="|", header=None, names=acq_colms)
data2016Q2= pd.read_csv('RawData/FannieMae_SingleFamily/Acquisition_2016Q2.txt', sep="|", header=None, names=acq_colms)
data2016Q3= pd.read_csv('RawData/FannieMae_SingleFamily/Acquisition_2016Q3.txt', sep="|", header=None, names=acq_colms)
data2016Q4= pd.read_csv('RawData/FannieMae_SingleFamily/Acquisition_2016Q4.txt', sep="|", header=None, names=acq_colms)
data2017Q1= pd.read_csv('RawData/FannieMae_SingleFamily/Acquisition_2017Q1.txt', sep="|", header=None, names=acq_colms)
data2017Q2= pd.read_csv('RawData/FannieMae_SingleFamily/Acquisition_2017Q2.txt', sep="|", header=None, names=acq_colms)
data2017Q3= pd.read_csv('RawData/FannieMae_SingleFamily/Acquisition_2017Q3.txt', sep="|", header=None, names=acq_colms)
data2017Q4= pd.read_csv('RawData/FannieMae_SingleFamily/Acquisition_2017Q4.txt', sep="|", header=None, names=acq_colms)
data2018Q1= pd.read_csv('RawData/FannieMae_SingleFamily/Acquisition_2018Q1.txt', sep="|", header=None, names=acq_colms)
data2018Q2= pd.read_csv('RawData/FannieMae_SingleFamily/Acquisition_2018Q2.txt', sep="|", header=None, names=acq_colms)
data2018Q3= pd.read_csv('RawData/FannieMae_SingleFamily/Acquisition_2018Q3.txt', sep="|", header=None, names=acq_colms)
```

```python
#See All Columns
pd.set_option('display.max_columns', 999)
data2015Q1
```

```python
#column data type
data2015Q1.dtypes
```

# Concatenate All Files

```python
#Concatenate Data
concatdata= pd.concat([data2015Q1,data2015Q2,data2015Q3,data2015Q4,data2016Q1,data2016Q2,data2016Q3,data2016Q4,
                      data2017Q1,data2017Q2,data2017Q3,data2017Q4,data2018Q1,data2018Q2,data2018Q3], join="inner")
```

```python
#Drop Any duplicate rows by LoanIds.
concatdata.drop_duplicates(['LoanIdentifier'])
```

### Scope: Single Family, Principal Residence and Purchase Only

```python
#Filter Single Family, Principal Residence and Purchase Only
acq_filter= concatdata[(concatdata.PropertyType=='SF')&(concatdata.OccupancyType=='P') & (concatdata.LoanPurpose=='P')] 

# Change First Time Home Buyer to Boolean
acq_filter['FirstTimeHomeBuyerIndicator'].replace(['N','Y'],[0,1], inplace=True)

#Fill NaN with 0 for Co-Borrower Credit Score
acq_filter['CoBorrowerCreditScoreAtOrigination'].fillna(0, inplace=True)

#Fill NaN with 0 for Primary Mortgage Insurance Percent (PMI)
acq_filter['PrimaryMortgageInsurancePercent'].fillna(0, inplace=True)

acq_filter.head()
```

# New DF

```python
#New DF Single Family, Only Purchases and Primary Residence
acq_clean= pd.DataFrame(
    acq_filter, columns=['LoanIdentifier','SellerName','OriginalInterestRate','OriginalUPB','OriginalLoanTerm','OriginationDate',
                       'FirstPaymentDate','OriginalLoanToValueLTV','PrimaryMortgageInsurancePercent','OriginalDebtToIncomeRatio','PropertyType','NumberofBorrowers','FirstTimeHomeBuyerIndicator',
                       'BorrowerCreditScoreAtOrigination', 'PropertyState','ZipCodeShort','CoBorrowerCreditScoreAtOrigination'])
acq_clean
```

```python
#Drop rows with NaN
acq_clean=acq_clean.dropna()
```

```python
acq_clean['FirstTimeHomeBuyerIndicator'].unique()
#Fannie Mae does not specify how to treat U so this field will become categorical
```

```python
#Reduce dtype size and chage objects

acq_clean['SellerName']=pd.Categorical(acq_clean['SellerName'], ordered= True)
acq_clean['OriginalInterestRate']=pd.to_numeric(acq_clean['OriginalInterestRate'], errors='coerce').astype(np.float32)
acq_clean['OriginalUPB']=pd.to_numeric(acq_clean['OriginalUPB'], errors='coerce').astype(np.int32)
acq_clean['OriginalLoanTerm']=pd.to_numeric(acq_clean['OriginalLoanTerm'], errors='coerce').astype(np.int32)
acq_clean['OriginationDate']=pd.to_datetime(acq_clean['OriginationDate'], format= '%m%Y',errors='ignore')  
acq_clean['FirstPaymentDate']=pd.to_datetime(acq_clean['FirstPaymentDate'], format= '%m%Y',errors='ignore')
acq_clean['OriginalLoanToValueLTV']=pd.to_numeric(acq_clean['OriginalLoanToValueLTV'], errors='coerce').astype(np.int32)
acq_clean['PrimaryMortgageInsurancePercent']=pd.to_numeric(acq_clean['PrimaryMortgageInsurancePercent'], errors='coerce').astype(np.int32)
acq_clean['OriginalDebtToIncomeRatio']=pd.to_numeric(acq_clean['OriginalDebtToIncomeRatio'], errors='coerce').round().astype(np.int32)
acq_clean['NumberofBorrowers']=pd.to_numeric(acq_clean['NumberofBorrowers'], errors='coerce').astype(np.int32)
#acq_clean['FirstTimeHomeBuyerIndicator']=pd.to_numeric(acq_clean['FirstTimeHomeBuyerIndicator'], errors='coerce').astype(np.int32)
acq_clean['BorrowerCreditScoreAtOrigination']=pd.to_numeric(acq_clean['BorrowerCreditScoreAtOrigination'], errors='coerce').round().astype(np.int32)
acq_clean['PropertyState']=pd.Categorical(acq_clean['PropertyState'], ordered= True)
acq_clean['ZipCodeShort']=pd.to_numeric(acq_clean['ZipCodeShort'], errors='coerce').astype(np.int32)
acq_clean['CoBorrowerCreditScoreAtOrigination']=pd.to_numeric(acq_clean['CoBorrowerCreditScoreAtOrigination'], errors='coerce').round().astype(np.int32)
```

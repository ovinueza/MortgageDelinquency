# Concatenate Years

```python
# per_data= pd.concat([data2015Q1, data2015Q2, data2015Q3, data2015Q4])
# per_data= pd.concat([data2016Q1, data2016Q2, data2016Q3, data2016Q4])
# per_data= pd.concat([data2017Q1, data2017Q2, data2017Q3, data2017Q4])
per_data= pd.concat([data2018Q1, data2018Q2, data2018Q3])
per_data.head()
```

```python
#Check Data Types
per_data.dtypes
```

```python
#Check NaNs
per_data.isna().sum()
```

```python
# If loan defaulted = 1, otherwise = 0
per_data['ForeclosureDate'].fillna(0, inplace=True)
per_data.loc[per_data.ForeclosureDate!=0, 'ForeclosureDate']=1
#Comfirm Values are 0,1
per_data['ForeclosureDate'].unique()
#Convert data type to interger
per_data['ForeclosureDate'] = per_data['ForeclosureDate'].astype(int)
#Count the number of records where Foreclosure =1
per_data[(per_data['ForeclosureDate']==1)].count()
```

```python
#Fill NaN with 0 for delinquency
per_data['CurrentLoanDelinquencyStatus'].fillna(0, inplace=True)
```

```python
#Check Foreclosed Values
#pd.set_option('display.max_rows',700)
per_data[(per_data.ForeclosureDate==1)]
```

```python
#Check number of foreclosures
per_data['ForeclosureDate'].sum()
```

```python
#Mark Delinquency as D when Foreclosure is True
mask= (per_data['ForeclosureDate']==1)
per_data['CurrentLoanDelinquencyStatus'][mask]='D'
```

```python
#Check Foreclosed and delinquency
per_data[(per_data.ForeclosureDate==1)]
#Check Delinquency = D
per_data[(per_data.CurrentLoanDelinquencyStatus=='D')]
```

```python
#Choose a loan from above kernel to see how default data plays out
per_data[(per_data.LoanIdentifier==768455311804)]
```

```python
#Check NANs
per_data.isna().sum()
```

```python
per_data['AdjustedMonthstoMaturity'].fillna(0, inplace=True)
```

```python
per_data.isnull().sum()
```

```python
#change dtypes for size
per_data['LoanAge']=pd.to_numeric(per_data['LoanAge'], errors='coerce').astype(np.int32)
per_data['RemainingMonthstoMaturity']=pd.to_numeric(per_data['RemainingMonthstoMaturity'], errors='coerce').astype(np.int32)
per_data['AdjustedMonthstoMaturity']=pd.to_numeric(per_data['AdjustedMonthstoMaturity'], errors='coerce').round().astype(np.int32)
per_data['MetropolitanStatisticalAreaMSA']=pd.to_numeric(per_data['MetropolitanStatisticalAreaMSA'], errors='coerce').astype(np.int32)
per_data['CurrentLoanDelinquencyStatus']=pd.Categorical(per_data['CurrentLoanDelinquencyStatus'], ordered=True)
```
